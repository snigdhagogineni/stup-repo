```python
#from descriptive statistics we observed high variance in data so we are performing min-max scaling to scale down entire data from 0-1
scaler=MinMaxScaler()
new_data=pd.DataFrame(scaler.fit_transform(new_data))
new_data.columns=['Hours Studied','Previous Scores','Performance Index']
```

```python
#capturing trget column in variable y and remaining columns in x variable
y=new_data['Performance Index']
new_data.drop('Performance Index',inplace=True,axis=1)
x=new_data
```

```python
#we will split x,y into train and test by 80:20 ratio
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
#checking size of all 4 sets of data
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# function to compute cost=(1/2m)(sum of (error*error)) where error is (predicted-actual) and predicted=(x*theta)
def cost_calculator(x, actual, theta):
  sample_count = len(x)
  #using np.dot() for matrix multiplication and np.sum() for summation
  cost = np.sum((x.dot(theta) - actual ) ** 2)/(2 * sample_count)
  return cost
```

```python
from sklearn.metrics.pairwise import pairwise_distances_chunked
# function to compute best set of parameters using gradient descent technique
def linear_regression_using_gradient_descent(x, actual, theta, learning_rate, iteration_count):
 all_iterations_cost = [0] * iteration_count
 sample_count = len(x)

 for iteration in range(iteration_count):
    #using dot() to multiply x and theta matrices to compute predicted values
    predicted= x.dot(theta)
    #error is difference between predicted and actual y values
    loss = predicted - actual
    #gradient is average of error multiplied by input matrix (input matrix has been transposed for matrix multiplication compatibility)
    gradient = x.T.dot(loss) / sample_count
    #updating theta with -learning rate*gradient
    theta = theta - learning_rate * gradient
    #computing cost of current x,y,theta values
    cost = cost_calculator(x, actual, theta)
    all_iterations_cost[iteration] = cost

 return theta, all_iterations_cost
```

```python
new_ones=np.ones((x_train.shape[0],1))#adding extra column of ones to x_train for matrix multiplication compatibility
x_train['intercept']=new_ones
```

```python
coefficients = np.zeros(x_train.shape[1])
learning_rate=0.1
number_of_iterations=10000
final_coefficients, all_iterations_cost = linear_regression_using_gradient_descent(x_train, y_train, coefficients,learning_rate,number_of_iterations)
print(final_coefficients)#printing 2 coefficients -one per column
print(all_iterations_cost)#keeping track of cost per iteration
```

```python
#generating iteration number vs cost function plot
pl.plot(all_iterations_cost)
pl.xlabel("Iteration number")
pl.ylabel("Cost function(Loss)")
pl.title("Iterations vs Cost function plot")
pl.show()
print('final cost after {} iterations is: {}'.format(number_of_iterations,all_iterations_cost[-1]))
```

```python
#from linear regression using gradient descent,the equation is Performance Index=(0.042)+(0.254)*Hours Studied+(0.668)*Previous Scores
```

```python
new_ones=np.ones((x_test.shape[0],1))#adding extra column of ones to x_test for matrix multiplication compatibility
x_test['intercept']=new_ones
```

```python
#using gradient descent model to predict on x_train (x_train*theta)-using np.dot() for matrix multiplication
predicted_y_train=x_train.dot(final_coefficients)
```

```python
mse=mean_squared_error(y_train,predicted_y_train)
print('Mean squared error for training dataset is:',mse)
print('Root Mean squared error for training dataset is:',math.sqrt(mse))
print('Mean absolute error for training dataset is:',mean_absolute_error(y_train,predicted_y_train))
print('coefficient of determination for training dataset is:',r2_score(y_train,predicted_y_train))
print('explained variance score for training dataset is:',explained_variance_score(y_train,predicted_y_train))
```

```python
#using gradient descent model to predict on x_test (x_test*theta)-using np.dot() for matrix multiplication
y_predicted=x_test.dot(final_coefficients)
```

```python
#evaluating testing performance based on mean absolute error,mean squared error,root mean squared error,R2 metrics
mse=mean_squared_error(y_test,y_predicted)
print('Mean squared error for training dataset is:',mse)
print('Root Mean squared error for training dataset is:',math.sqrt(mse))
print('Mean absolute error for training dataset is:',mean_absolute_error(y_test,y_predicted))
print('coefficient of determination for training dataset is:',r2_score(y_test,y_predicted))
print('explained variance score for training dataset is:',explained_variance_score(y_test,y_predicted))
```
